```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
iris["species"].value_counts()
```

```python
i_groups=iris.groupby(iris.species)   #iris  
```

```python
i_groups.sum()
```

#### agg

```python
def peak_to_peak_ratio(x):
    return x.max()/x.min()    #      
```

```python
i_groups.agg(peak_to_peak_ratio)   
# in-[57]     ()    
# peak_to_peak_ratio  
    
```

```python
i_groups.apply(peak_to_peak_ratio)    
```

# apply 

```python
def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3] #    
```

```python
iris.groupby(iris.species).apply(top3_petal_length)
```

```python
iris.groupby(iris.species).agg(top3_petal_length)


#   #   (agg  )
```

apply 

```python
def q3cut(s):
    return pd.qcut(s,3, labels=["","",""]).astype(str)
```

```python
iris.groupby(iris.species).petal_length.apply(q3cut)
```

```python
iris["petal_length.class"]=iris.groupby(iris.species).petal_length.apply(q3cut)

iris.head(10)
iris.tail(10)
```
