### 열 또는 행에 동일한 연산 반복 적용할 때 : apply() 함수
- apply() 함수는 DataFrame의 행이나 열에 복잡한 연산을 vectorizing할 수 있게 해주는 함수로 매우 많이 활용되는 함수임
- 동일한 연산을 모든열에 혹은 모든 행에 반복 적용하고자 할때 사용

- apply(반복적용할 함수, axis=0/1)
    - 0 : 열마다 반복
    - 1 : 행마다 반복 
    - 생략시 기본값 : 0


```python
#예제 df 생성
df3 = pd.DataFrame({
    'a':[1,3,4,3,4],
    'b':[2,3,1,4,5],
    'c':[1,5,2,4,4]
})
df3
```

```python
# df3의 각 열에 대해 np.sum 이라는 함수를 반복 적용하는 코드를 생성하시오

df3.apply(np.sum)
```

```python
# df3의 각 행에 대해 np.sum 이라는 함수를 반복 적용하는 코드를 생성하시오

df3.apply(np.sum,1)
```

```python
df3.sum(axis=0)
# 위 코드와 같은 결과를 반환
```

```python
df3.apply(diff,0)
```

```python
# apply 함수를 이용하여 위에서 생성한 lambda diff를 df3의 모든 행에 반복 적용하여
# 모든 행의 최대값과 최소값의 차이를 구하시오
```

```python
df3.apply(diff,1)
```

#### 관측 데이터 값을 범주형(카테고리) 값으로 변환
- 값의 크기를 기준으로 하여 카테고리 값으로 변환하고 싶을 때
    - cut(date, bins, label)
    - data : 구간을 나눌 실제 관측값,  bins: 구간 경계 값 , label: 카테고리 값
    

```python
ages=[0,0.5,1,4,5,10,20,21,25,30,34,36,40,41,45,46,50,51,52,55,100]
```

```python

#label : 카테고리 명
labels=['영유아','미성년자','청년','중년','장년','노년']


# bins: 구가 경계값 설정
bins=[0,4,15,25,35,60,100]


# 0<영유아<=4
# 4<미성년자<=15
```

```python
# 함수 적용해서 카테고리 생성
cats=pd.cut(ages, bins=bins,labels=labels)
cats
```

```python
list(cats)    #nan-->0살   지정된 범위가 없어서 판단못하니 nan으로
```

# 리스트 내포 연산


#### 리스트 내포 for문 :문법
- [표현식(연산식) for 항목 in 반복가능객체 if 조건문]
- if 조건문은 생략 가능하다.
- 반복가능객체 : 리스트, 튜플,딕셔너리,range()등


```python
a=[1,2,3,4]
# 위 a 리스트의 각 원소에 대해 2배한 원소값을 만들고 result라는 변수에 저장하시오

#[1,2,3,4]*2  #[1,2,3,4,1,2,3,4]


result=[]
for i in a:
    result.append(i*2)
    
result

```

```python
# 내포 for 문 사용

result=[i*2 for i in a]
result
```
