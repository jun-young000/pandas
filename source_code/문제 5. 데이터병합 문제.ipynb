### 연습문제 1
- 두개의 데이터프레임을 만들고 merge 명령을 병합한다.
- 단, 데이터 프레임은 다음 조건을 만족해야 한다.
    1. 각각 5X5 이상의 크기를 가진다.
    2. 공통열을 하나이상 갖는다. 
        - 다만 공통열의 이름은 서로 다르게 할 것
    3. merge의 경우를 inner, outer, left, right 4개의 형태로 출력할 것
    4. 지정된 인덱스와 컬럼명을 갖는다. 

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '대구'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '강릉']]
df1 = pd.DataFrame(data, 
                   index=['홍길동', '이몽룡', '성춘향','변학도','김연아'], 
                   columns=["나이", "몸무게", "키", "성별" ,"주소"])
df1
```

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '울산'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '광주']]
df2 = pd.DataFrame(data,
                   index=[1,2,3,4,5],
                   columns=["나이", "몸무게", "키", "성별" ,"지역"])
df2
```

```python
pd.merge(df1,df2, how="outer")
pd.merge(df1,df2, how="inner")
pd.merge(df1,df2, how="left")
pd.merge(df1,df2, how="right")

# 복수의 키값이 존재한다
```

```python
### 이름이 다르지만 동일한 데이터 영역이다. 이상태에서 이 다른 값을 키로 잡으려면 _on을 쓴다

pd.merge(df1,df2, left_on='주소', right_on='지역')
```

```python
pd.merge(df1,df2, how="outer",left_on='주소', right_on='지역')
```

```python
pd.merge(df1,df2,how="left", left_on='주소', right_on='지역')
```

```python
pd.merge(df1,df2,how="right", left_on='주소', right_on='지역')
```

### 연습문제 3
- 어느회사의 전반기(1월-6월)실적을 나타내는 데이터프레임과 후반기(7월-12월)실적을 나타내는  데이터 프레임을 작성한다

- 실적 정보는 "매출","비용","이익"으로 이루어진다.
- '매출','비용'은 임의의 데이터를  사용하고
- 이익은 이익=매출-비용의  가공필드로 생성한다
    
- 또한 1년간의 총 실적을 마지막 열로 덧붙인다.

![](데이터병합1.PNG)

![](데이터병합2.PNG)

```python
# 데이터프레임 생성

month = [str(i)+'월' for i in range(1,13)]
month

np.random.seed(3)

#매출, 비용 data random  생성
data={}
for i in month[0:6] :
    data[i] = np.random.randint(1,100,2)
data 

#데이터프레임 생성
df_score3 = pd.DataFrame(data,
                        index=['매출','비용'])
display(df_score3)

#가공 행 이익 생성 코드 작성
df_score3.loc["이익"]=df_score3.loc["매출"]-df_score3.loc["비용"]

display(df_score3)

#매출, 비용 data random  생성
data={}
for i in month[6:12] :
    data[i] = np.random.randint(1,100,2)
data 

#데이터프레임 생성
df_score4 = pd.DataFrame(data,
                        index=['매출','비용'])
display(df_score4)

#가공 행 이익 생성 코드 작성

df_score4.loc["이익"]=df_score4.loc["매출"]-df_score4.loc["비용"]
df_score4
```

### 연습문제 3_1

```python
df_total=pd.concat([df_score3, df_score4], axis=1)
df_total["총실적"]=df_total.sum(axis=1)
df_total
```
