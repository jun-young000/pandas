## 카이제곱 검정 개요


- 목적 : 두 범주형 변수가 서로 독립적인지 검정

    - 귀무가설: 두 변수가 서로 독립이다.
    
    - 대립가설: 두 변수가 서로 종속이다.
    
    
- 교차 테이블(분할표)로 시각화

### 교차 테이블(분할표)

- 두 변수가 취할 수 있는 값의 조합의 출현 빈도를 나타냄


- 기대값과 실제값의 차이가 클수록 통계량이 커지게 됨 통계량이 커질수록 P-VALUE가 감소함


- 귀무가설이 기각될 가능성이 높아진다.

### 분할표 작성
- pd.crosstab()

     - 범주형 변수로 되어있는 요인(factors)별로 교차분석(cross tabulations) 해서, 행, 열 요인 기준 별로 빈도를 세어서 도수분포표(frequency table), 교차표(contingency table) 를 만들어줌


```python
# 교차 테이블 생성(분할표)


cross_t = pd.crosstab(df["smoke"], df["disease"])

cross_t


type(cross_t)
```

```python
obs=cross_t.values # 값만 넘겨줘야 하니깐 values 만 하면됨
```

```python
obs                 # df에서 값을 뽑았으니 이차원 행렬
```

```python
statistics,pvalue,dof,expected = chi2_contingency(obs,correction=False)
# 통계량, pvalue, 자유도, 기대값
# 기대값의 기대도수가 5를 초과하면 False


pvalue
# 0.026855610352114926 나오므로 귀무가설이 기각=흡연과 폐암 유무는 연관성이 있다.
# 0.05를 가 기준
```

```python
statistics
```

```python
pvalue
```

```python
dof
```

```python
expected
```

```python
# 기대값  

pd.DataFrame(expected, columns=cross_t.columns, index=cross_t.index)

# 기대값을 df로 바꿈
```

```python
# 기대값이 전부다 5를 넘는다~!!!!!!!!!!!!
```

```python
statistics
```

```python

###### 기대도수를 5미만으로 조정하면(자유도  correction=True )
# pvalue가 0.06457983327434368를 넘으므로 귀무가설은 성립
statistics,pvalue,dof,expected = chi2_contingency(obs)
pvalue
pd.DataFrame(expected, columns=cross_t.columns , index=cross_t.index)
```

```python
obs=cross_t.values
obs
```

```python
chi2_contingency(obs,correction=False)
# pvalue가 0.001892528960792166 이므로 귀무가설 기각
# 만족도와 성별은 서로 독립적이지 않다.
```
