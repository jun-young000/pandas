# 대표적인 연속형 확률 분포


- 연속형 확률 분포
    - 정규분포
    - t_분포
    - 카이제곱분포
    - F-분포
    
    


- 정규분포(normal distribution)
        - 대표본 모집단의 합, 평균 추론 시 활용
- t-분포
        - 소표본 모집단 평균 추론 시 활용
        - 선형모형 회귀계수 추론(종속변수 정규분포 가정)시 활용
- 카이제곱(χ2)분포
        - 모집단 분산 추론 시(데이터 정규분포 가정) 활용
        - 카이제곱 검정 시 활용
- F-분포
        - 두 모집단 분산 차이 비교 시 활용(데이터 정규분포 가정)
        - 분산분석 시 활용
        - (설명하는 변동/설명하지 못하는 변동)이 F분포를 따름


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, integrate
from scipy.optimize import minimize_scalar

%precision 3
%matplotlib inline
```

```python
linestyles = ['-', '--', ':']

def E(X, g=lambda x: x):
    x_range, f = X
    def integrand(x):
        return g(x) * f(x)
    return integrate.quad(integrand, -np.inf, np.inf)[0]

def V(X, g=lambda x: x):
    x_range, f = X
    mean = E(X, g)
    def integrand(x):
        return (g(x) - mean) ** 2 * f(x)
    return integrate.quad(integrand, -np.inf, np.inf)[0]

def check_prob(X):
    x_range, f = X
    f_min = minimize_scalar(f).fun
    assert f_min >= 0, 'density function is minus value'
    prob_sum = np.round(integrate.quad(f, -np.inf, np.inf)[0], 6)
    assert prob_sum == 1, f'sum of probability is {prob_sum}'
    print(f'expected vaue {E(X):.3f}')
    print(f'variance {V(X):.3f}')
    
def plot_prob(X, x_min, x_max):
    x_range, f = X
    def F(x):
        return integrate.quad(f, -np.inf, x)[0]

    xs = np.linspace(x_min, x_max, 100)

    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111)
    ax.plot(xs, [f(x) for x in xs],
            label='f(x)', color='gray')
    ax.plot(xs, [F(x) for x in xs],
            label='F(x)', ls='--', color='gray')

    ax.legend()
    plt.show()

```

## 정규 분포

- 가장 대표적인 분포
- 가우시안(Gaussian distribution)분포 라고도 불리움
- 연속형 확률변수에 대한 분포
- 평균(μ)과 표준편차(σ)로 모양이 결정
- 좌우대칭, 종모양
- 중앙 한 점이 뽀족
- 평균=중앙값=최빈값
- 자연계ㅡ 사회현상에 많이 나타난다.





### 정규분포의 확률 밀도 함수 f(x)





- 확률밀도함수




```python
def N(mu, sigma):  #mu: 평균 , sigma 표준편차
    x_range=[-np.inf, np.inf]
    
    def f(x):
        return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-(x-mu)**2/(2*sigma**2))
    return x_range,f
```

```python
# 평균이 2이고 표준편차가 0.5인 정규분포를 따르는 확률변수 x정의

mu=2
sigma=0.5

X=N(mu,sigma)
```

```python
# 기대값과 분산을 출력해 줌
# 해당 함수는 3번줄에 정의해놓음
check_prob(X)  
```

```python
# 정규분포를 따르는 확률변수 X의 그래프


plot_prob(X,0,4)
```

-남자 고등학생 키의  평균이 170CM이고 표준편차가 5CM라면 우연히 만난 남자 고등학생의 키는 N(170,5^2)를 따른다고 할수 있다.

### scipy.stats 사용해서 확인


- stats.norm(평균,표준편차):정규분포를 따른 확률 변수를 객체 인스턴스로 반환

```python
# 기대값이 2이고 표준편차가 0.5인 정규분포를 따르는 확률변수 rv를 생성
rv=stats.norm(2,0.5)
```

```python
# rv의 확률변수의 평균과 분산을 계산
# mean(), var()

rv.mean()
rv.var()
```

```python
# 확률밀도함수:pdf()
rv.pdf(2)

```

```python
# rv 확률변수의 분포를 그래프로 확인
xx=np.linspace(-2,6,100)
pdf=rv.pdf(xx)
plt.plot(xx,pdf)
# xx
# pdf
plt.title("확률밀도함수 ")
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.show()
```

- 확률 변수의 누적 확률 분포 값: cdf메서드로 계산
- 누적분포함수: p(X <= x)를 꼐산하는 함수
- norm.cdf(x)

```python
# 확률변수 rv가 1.7보다 작은값이 되는 확률
rv.cdf(1.7)
```

-상위100 a%지점



- norm.isf(위치값) : 상위 위치값 % 지점을 반환

```python

xx=np.linspace(-2,6,100)
pdf=rv.pdf(xx)
plt.plot(xx,pdf)
# 상위 30% 지점값 반환
rv.isf(0.3)
```

```python
xx=np.linspace(-2,6,100)
pdf=rv.pdf(xx)
plt.plot(xx,pdf)
rv.isf(0.95)
```

-norm.interval(구간비율)
    
    -norm.interval(0.9): 90% 구간 지검을 반환- 왼쪽과 오른쪽 끝에 5%씩 남김

```python
# 90% 구간
xx = np.linspace(-2, 6, 1000)
pdf = rv.pdf(xx)
plt.plot(xx,pdf)
rv.interval(0.9)
```

```python
# 왼쪽과 오른쪽 긑에서 5% 씩 남김

rv.isf(0.95), rv.isf(0.05)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

xs = np.linspace(-5, 5, 100)
params = [(0, 1), (0, 2), (1, 1)]
for param, ls in zip(params, linestyles):
    mu, sigma = param
    rv = stats.norm(mu, sigma)
    ax.plot(xs, rv.pdf(xs),
            label=f'N({mu}, {sigma**2})', ls=ls, color='gray')
ax.legend()

plt.show()
```

##  카이제곱 분포



- z는 정규확률변수, 자유도는 정규분포의 갯수(n)


```python
n=10

rv=stats.norm(0,1)

sample_size=int(1e6) #100000



#표준 정규분포로부터 10*100만 사이즈로 무작위 data(난수)를 추출
Za_sample=rv.rvs((n,sample_size))
Za_sample


```

```python
chi_sample=np.sum(Za_sample**2, axis=0)

len(chi_sample)
```

```python
chi_sample
```

### 카이 제곱 분포를 따르는 확률 변수 생성: stats.chi2(n)

```python
n=10
rv_true=stats.chi2(n)

xs=np.linspace(0,30,100)
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111)
# sample 데이터 카이제곱 분포 계산 데이터를 이용한 히스토그램
ax.hist(chi_sample, bins=100, density=True , alpha=0.5, label="chi_sample")

ax.plot(xs,rv_true.pdf(xs),label=f'chif({n})',color="gray")
```

```python
#자유도 값을 변경해가면서 그래프 모양 확인
```

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111)

xs=np.linspace(0,20,500)

for n, ls in zip([3,5,10],linestyles):
    rv=stats.chi2(n)
    ax.plot(xs,rv.pdf(xs),label=f'chif({n})',ls=ls,color="gray")
    
ax.legend()
plt.show()
    
```

```python
rv=stats.chi2(5)  #자유도가 5일떄
rv.isf(0.05)   #상위 5%
rv.isf(0.95)   #상위 95%
```

## t분포- 정규분포에서 모평균의 구간추청등에 사용되는 확률분포
- stats.t(n)       # n은 자유도

- 좌우 대칭인 분포
- 표준 정규분포보다 양쪽 끝이  더 두껍다.
- 자유도가 커지면 표준 정규분포에 가까워 진다.


```python
n=10
rv1=stats.norm()
rv2=stats.chi2(n)


sample_size=int(1e6)


Z_sample = rv1.rvs(sample_size)
Y_sample = rv2.rvs(sample_size)

t_sample = Z_sample /  np.sqrt(Y_sample/n)



len(t_sample)
```

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111)



# t 확률변수 생성


rv=stats.t(n)
xs=np.linspace(-3,3,100)

# 직접 계산해서 생성한 t_sample의 히스토그램

ax.hist(t_sample,bins=100, range=(-3,3),  density=True , alpha=0.5, label="t_sample")

#확률 밀도 함수 이용해 xs의 대응하는 y값 계산하고 그래프 그리기

ax.plot(xs, rv.pdf(xs), label=f'chif({n})',color="gray")    

ax.legend()
ax.set_xlim(-3, 3)
plt.show()

```

```python
# 정규 분포와 매우 유사함
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

xs = np.linspace(-3, 3, 100)
for n, ls in zip([3, 5, 10], linestyles):
    rv = stats.t(n)
    ax.plot(xs, rv.pdf(xs),
            label=f't({n})', ls=ls, color='gray')
rv = stats.norm()
ax.plot(xs, rv.pdf(xs), label='N(0, 1)')
    
ax.legend()
plt.show()
```

```python
rv=stats.t(5)
rv.isf(0.05)
```

## F분포-분산분석 등에서 사용되는 확률 분포


```python
n1=5
n2=10


rv1=stats.chi2(n1)
rv2=stats.chi2(n2)

sample_size=int(1e6)

Y1=rv1.rvs(sample_size)
Y2=rv2.rvs(sample_size)


f_sample=(Y1/n1)/(Y2/n2)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)



# 자유도가 n1,n2인 F분포를 따르는 확률변수 rv를 생성
rv=stats.f(n1,n2)
xs=np.linspace(0,6,200)[1:]

ax.plot(xs,rv.pdf(xs),label=f'F({n1}, {n2})', color='gray')


ax.hist(f_sample, bins=100, range=(0, 6),
        density=True, alpha=0.5, label='f_sample')


ax.legend()
ax.set_xlim(0, 6)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

xs = np.linspace(0, 6, 200)[1:]
for n1, ls in zip([3, 5, 10], linestyles):
    rv = stats.f(n1, 10)
    ax.plot(xs, rv.pdf(xs),
            label=f'F({n1}, 10)', ls=ls, color='gray')
    
ax.legend()
plt.show()
```
