# Ïó∞ÏÜçÌòï ÌôïÎ•†Î≥ÄÏàòÏùò ÌôïÎ•†Ïùò ÏÑ±Ïßà


-Ï≤´Î≤àÏß∏ ÏÑ±Ïßà: Ìï®Ïàò f(x)Îäî 0Î≥¥Îã§ Í∞ôÍ±∞ÎÇò Ïª§ÏïºÌïúÎã§.
    
    - ÏúÑ ÏÑ±ÏßàÏùÑ ÌôïÏù∏ÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî minimize_scalar Ìï®Ïàò ÏÇ¨Ïö©
    
    - minimize_scalar(Ìï®Ïàò): Ìï®ÏàòÎ•º Ïã§ÌñâÌïú Í≤∞Í≥ºÍ∞íÏù¥ ÏµúÏÜåÍ∞Ä ÎêòÎäî xÍ∞íÍ≥º ÏµúÏÜåÍ≤∞Í≥º yÎ•º Î∞òÌôò
    

```python
from scipy.optimize import  minimize_scalar
def func(x):
    return (x-1.5)**2+0.5

res=minimize_scalar(func)
res.x  # ÏµúÏÜåÍ∞íÏùÑ ÎßåÎì§Ïñ¥ÎÇ∏ xÍ∞í
res.fun  # Ìï®ÏàòÍ∞Ä ÎßåÎì§Ïñ¥ ÎÇºÏàò ÏûáÎäî ÏµúÏÜåÍ∞í
func(1.5)
```

```python
# ÏúÑÏóêÏÑú ÏÉùÏÑ±Ìï¥ ÎÜìÏùÄ Î∂àÍ≥µÏ†ï Î£∞Î†õÏùò ÌôïÎ•†Ìï®Ïàò
def f(x):
    if x_range[0] <= x <= x_range[1]:
        return 2 * x
    else:
        return 0
```

```python
from scipy.optimize import  minimize_scalar
res=minimize_scalar(f)
res.x
res.fun   # 0Ïù¥ Î∞òÌôò Îê¨ÏúºÎØÄÎ°ú Ï≤´Î≤àÏß∏ ÌôïÎ•†Ïùò ÏÑ±ÏßàÏùÑ ÎßåÏ°±
```

- ÏúÑ Ïó∞ÏÇ∞Ïùò ÌôïÏù∏ÏúºÎ°ú Ïù∏Ìï¥ f(x)>=0Ïùò Ïó∞ÏÜçÌòï ÌôïÎ•†Î≥ÄÏàòÏùò ÏÑ±ÏßàÏù¥ ÎßåÏ°± Îê®

- ÎëêÎ≤àÏß∏ f(x)Î•º Ôºç‚àû Î∂ÄÌÑ∞ ‚àû ÍπåÏßÄ Ï†ÅÎ∂ÑÌïú Í≤∞Í≥ºÍ∞Ä 1Ïù¥ÎùºÎäî Í≤ÉÏùÄ ÏúÑ Í∑∏Î¶ºÏóêÏÑú ÏÇºÍ∞ÅÌòïÏùò Î©¥Ï†ÅÏù¥ 1Ïù¥ ÎêòÎäî Í≤ÉÍ≥º Í∞ôÎã§
    - Ïù¥ ÏÇºÍ∞ÅÌòïÏùÄ Î∞ëÎ≥ÄÏùò Í∏∏Ïù¥Í∞Ä 1, ÎÜíÏù¥Í∞Ä 2 Ïù¥ÎØÄÎ°ú Î©¥Ï†ÅÏù¥ 1Ïù¥ÎùºÎäî Í≤ÉÏùÑ Í∞ÑÎã®Ìûà Ïïå Ïàò ÏûàÏùå


### ÌôïÎ•†Ïùò ÎëêÎ≤àÏ®∞ ÏÑ±ÏßàÏùÑ ÏàòÏπòÏ†ÅÎ∂Ñ Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ ÎßåÏ°±ÌïòÎäîÏßÄ ÌôïÏù∏

#### quad(): ÏàòÏπòÏ†ÅÎ∂Ñ Ìï®Ïàò

- **ÏàòÏπòÏ†ÅÎ∂Ñ(numerical integration)**ÏùÄ Ìï®ÏàòÎ•º ÏïÑÏ£º ÏûëÏùÄ Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÏñ¥ Ïã§Ï†ú Î©¥Ï†ÅÏùÑ Í≥ÑÏÇ∞Ìï®ÏúºÎ°úÏç® Ï†ïÏ†ÅÎ∂ÑÏùò Í∞íÏùÑ Íµ¨ÌïòÎäî Î∞©Î≤ïÏù¥Îã§. 
- ScipyÏùò integrate ÏÑúÎ∏åÌå®ÌÇ§ÏßÄÏùò quad Î™ÖÎ†πÏúºÎ°ú ÏàòÏπòÏ†ÅÎ∂ÑÏùÑ Ìï† Ïàò ÏûàÎã§.

- Ï≤´ Î≤àÏß∏ Ïù∏ÏàòÎäî ÌîºÏ†ÅÎ∂ÑÌï®Ïàò„ÄÅÎëê Î≤àÏß∏ Ïù∏ÏàòÏôÄ ÏÑ∏ Î≤àÏß∏ Ïù∏ÏàòÎäî Ï†ÅÎ∂Ñ Î≤îÏúÑ
- Ï≤´Î≤àÏß∏ Î∞òÌôòÍ∞íÏù¥ ÏàòÏπò Ï†ÅÎ∂ÑÏúºÎ°ú ÏñªÏñ¥ÏßÑ Í≤∞Í≥ºÏù¥Î©∞, ÎëêÎ≤àÏß∏Ïùò Í∞íÏùÄ Ï∂îÏ†ï Ïò§Ï∞®


```python
# ÎëêÎ≤àÏß∏ ÏÑ±ÏßàÏùÑ Ï†ÅÎ∂Ñ Í≥ÑÏÇ∞ÏúºÎ°ú ÌôïÏù∏
# Ôºç‚àû Î∂ÄÌÑ∞ ‚àûÎäî np.infÎ°ú ÌëúÌòÑÌï† Ïàò ÏûàÏùå=>np.inf: Î¨¥ÌïúÎåÄ  - np.inf: ÎßàÏù¥ÎÑàÏä§Î¨¥ÌïúÎåÄ

integrate.quad(f,-np.inf,np.inf)


# f(x)Î•º -Î¨¥ÌïúÎåÄÎ∂ÄÌÑ∞ +Î¨¥ÌïúÎåÄÍπåÏßÄ Ï†ÅÎ∂ÑÌïú Í≤∞Í≥ºÍ∞Ä 1Ïù¥Ïñ¥ÏïºÌï®- ÎëêÎ≤àÏß∏ ÌôïÎ•†Ïùò ÏÑ±ÏßàÏùÑ ÎßåÏ°±
```

- ÏúÑ ÏÇ¨Îã§Î¶¨Íº¥ ÏòÅÏó≠ÏùÑ Ï†ÅÎ∂ÑÏúºÎ°ú Íµ¨ÌïòÎäî Ïãù










```python
# 2xÎäî Ìï®Ïàò f(x)Í∞Ä Íµ¨ÌòÑÌïòÍ≥† ÏûàÏùå
integrate.quad(f, 0.4, 0.6)

# Ï†ÅÎ∂ÑÍ∞íÏù¥ 0.200 Ïù¥ÎØÄÎ°ú 0.4ÏóêÏÑú 0.6Ïù¥ ÎÇòÌÉÄÎÇ† ÌôïÎ•†ÏùÄ  0.2
```

## ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò
- xÍ∞Ä xÏù¥ÌïòÍ∞Ä Îê†ÎïåÏùò ÌôïÎ•†ÏùÑ Î∞òÌôòÌïòÎäî Ìï®Ïàò

```python
# Î∂ÑÌè¨Ìï®Ïàò Íµ¨ÌòÑ

def F(x):
    return integrate.quad(f,- np.inf,x )[0]  # [0] Ï†ÅÎ∂ÑÍ∞íÏù¥ Îì§Ïñ¥ÏûàÏùå
```

- Î£∞Î†õ 0.4ÏóêÏÑú 0.6ÏÇ¨Ïù¥Ïùò Í∞íÏùÑ Ï∑®Ìôú ÌôïÎ•†



```python
F(0.6) - F(0.4) 
```

- ÏúÑ Î∂ÑÌè¨ Ìï®Ïàò F(x)Î•º Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î†§Î≥¥Î©¥ Î∞òÎìúÏãú Îã®Ï°∞ Ï¶ùÍ∞ÄÌï®Ïàò(xÍ∞Ä Ï¶ùÍ∞ÄÌï† Îïå yÎäî Í∞êÏÜåÌïòÏßÄ ÏïäÎäî Ìï®Ïàò)Í∞Ä Îê®


```python
# Îã®Ï°∞ Ï¶ùÍ∞Ä Ìï®ÏàòÌôïÏù∏

xs = np.linspace(x_range[0], x_range[1], 100)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.plot(xs, [F(x) for x in xs], label='F(x)', color='gray')
ax.hlines(0, -0.1, 1.1, alpha=0.3)
ax.vlines(0, -0.1, 1.1, alpha=0.3)
ax.vlines(xs.max(), 0, 1, linestyles=':', color='gray')

ax.set_xticks(np.arange(-0.1, 1.2, 0.1))
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)
ax.legend()

plt.show()
```

### Ïó∞ÏÜçÌòï ÌôïÎ•† Î∂ÑÌè¨Ïùò ÏßÄÌëú

#### ÌèâÍ∑†-Í∏∞ÎåÄ Í∞í



```python
# ÏúÑÏóêÏÑú ÏÉùÏÑ±Ìï¥ ÎÜìÏùÄ Î∂àÍ≥µÏ†ï Î£∞Î†õÏùò ÌôïÎ•†Ìï®Ïàò
def f(x):
    if x_range[0] <= x <= x_range[1]:
        return 2 * x
    else:
        return 0
```

```python
def integrand(x): # (x(fx)) Í≥ÑÏÇ∞ Í≤∞Í≥ºÎ•º Î∞òÌôòÌïòÎäî Ìï®Ïàò
    return x*f(x)

integrate.quad(integrand, -np.inf, np.inf)[0]
```

```python
# ÌôïÎ•†Î≥ÄÏàò xÎ•º Ï†ïÏùò

X=[x_range,f]
```

```python
def E(X, g=lambda x: x) :
    x_range, f = X
    def integrand(x) : # (x(fx)) Í≥ÑÏÇ∞ Í≤∞Í≥ºÎ•º Î∞òÌôòÌïòÎäî Ìï®Ïàò
        return g(x)* f(x)
    return integrate.quad(integrand, -np.inf, np.inf)[0]
```

```python
E(X)                   # ÏΩîÎìú Ïû¨ÌôïÏù∏
```

```python
E(X,g=lambda x : 2*x+3)
```

```python
2*E(X)+3
```

### Î∂ÑÏÇ∞

- Œº Îäî ÌôïÎ•†Î≥ÄÏàò XÏùò Í∏∞ÎåÄÍ∞í



```python
# ÌôïÎ•†Î≥ÄÏàò XÏóê ÎåÄÌïú Î∂ÑÏÇ∞Í≥ÑÏÇ∞


mean = E(X)  #ÌôïÎ•†Î≥ÄÏàò XÏóê ÎåÄÌïú Í∏∞ÎåÄÍ∞í


def integrand(x):
    return (x-mean)**2*f(x)
integrate.quad(integrand, -np.inf, np.inf)[0]
```
