#### 쌍체 표본 t검정 

- x의 변화에 따라 y의 값이 결정
- x에 변화를 가했을 때 그 x가 변화되어서 추출된 값을 검정(효과가 있는지)
- A반 학생들이 보충수업 후 본 국어 시험점수를 봤을 때 보충수업의 효과가 있는지


- 가설설정- 데이터 정규성 검정- Ttest- 결론




#### 선행 조건

- 실험전(x)과 실험후(y)의 데이터는 정규분포를 따르지 않아도 됨.


- 측정값의 차이는 정규성을 갖고 있어야 함



```python
import pandas as pd
from scipy.stats import *
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
df=pd.read_csv("C:/Users\Administrator/Desktop/coding/python/muticam/jupyter note/data/다이어트약_효과검증.csv")

df.head()  
df.tail()
```

```python
before=df["다이어트전(kg)"]
after=df["다이어트후(kg)"]
```

```python
# 정규성 검정: 두값의 차이가 정규분포를 따르는지 확인

kstest(after-before, "norm")


# pvalue 가 거의 0에 가까운 수치(지수가 -면 소수 이하로 내려간다.)
```

```python
# 정규성 검증을 kstest말고 다른식으로도 함.

# from scipy.stats import shapiro  (대표적인 정규성 검정 함수, 말그래로 정규성만 검증.정규분포만 확인한다는소리)

# 귀무가설:데이터는 정규분포이다.
# 대립가설:데이터는 정규분포가 아니다

shapiro(after-before)


# pvalue 가 0.05보다 크므로 귀무가설 채택
```

#### 대응(쌍체)표본  t 검정 함수

- scipy.stats.ttest_rel(a, b, axis=0, nan_policy='propagate', alternative='two-sided')
    - nan_policy {‘propagate’:nan을 반환, ‘raise’:오류발생, ‘omit’:무시} : NaN 처리 방법
    - alternative='two-sided'|'less'|'greater' : 양측검정|왼쪽단측|오른쪽단측


```python
# 귀무가설: 대응 표본 두 집단의 평균은 같다- 다이어트 약의 효과는 없다.
# 대립가설: 대응 표본 두 집단의 평균은 다르다- 다이어트 약의 효과가 있다.
#  pvalue 가 0에 가까우므로 귀무가설은 기각
# 통계량은 확실한 양의 효과가 있다.-다이어트 약의 효과가 있다.

ttest_rel(before, after)
```
